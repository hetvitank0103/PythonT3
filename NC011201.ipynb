```python
%%writefile findindex.py
def find(a,b):
    count=0
    for i in a:
        count+=1
        if i==b:
            break
    return count
def rfind(a,b):
    count=0
    for i in a:
        count+=1
    c=a[::-1]
    countb=0
    for i in c:
        countb+=1
        if i==b:
            break
    return count-countb+1
a="sdkjsdkasd sadkjkdjklas lkjlkjkldfsj ldkjflkdfs"
b="d"
```
