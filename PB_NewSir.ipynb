```python
#PB 535
class WordPlay:
    def __init__(self):
        self.L=eval(input('Enter list of words: '))
    def words_with_length(self,length):
        L=[]
        for i in self.L:
            if len(i)==length:
                L.append(i)
            return L
    def ends_with(self,a):
        L=[]
        for i in self.L:
            if i.endswith(a):
                L.append(i)
        return L
    def starts_with(self,a):
        L=[]
        for i in self.L:
            if i.startswith(a):
                L.append(i)
        return L
    def palindromes(self):
        L=[]
        for i in self.L:
            if i==i[::-1]:
                L.append(i)
        return L
    def only(self,a):
        L=[]
        for i in self.L:
            if set(i) ==set(a):
                L.append(i)
        return L
    def avoids(self,a):
        L=[]
        for i in self.L:
            if set(i) & set(a) == set():
                L.append(i)
        return L
        
w= WordPlay()
print(w.words_with_length(5))
print(w.starts_with('a'))
print(w.ends_with('d'))
print(w.palindromes())
print(w.only('bna'))
print(w.avoids('amkd'))
```

```python
#PB 533
class studentmgmt:
    students=[]
    def __init__(self,n,r,m1,m2):
        self.name=n
        self.rollno=r
        self.marks1=m1
        self.marks2=m2
        studentmgmt.students.append(self)
    def accept(self):
        n=input("Enter name: ")
        r=int(input("Enter roll no: "))
        m1=int(input("Enter marks 1: "))
        m2=int(input("Enter marks 2: "))
        studentmgmt(n,r,m1,m2)
    def display(self):
        for i in studentmgmt.students:
            print(f"Name: {i.name}")
            print(f"Roll no: {i.rollno}")
            print(f"Marks 1: {i.marks1}")
            print(f"Marks 2: {i.marks2}")
    def search(self):
        r=int(input("enter rollno to search: "))
        for i in studentmgmt.students:
            if i.rollno==r:
                print(f"Name: {i.name}")
                print(f"Roll no: {i.rollno}")
                print(f"Marks 1: {i.marks1}")
                print(f"Marks 2: {i.marks2}")
                return
    def delete(self):
        r=int(input("enter rollno to deleter: "))
        for i in studentmgmt.students:
            if i.rollno==r:
                studentmgmt.students.remove(i)
                return
            print(f"student with roll no {r} not found")
    def update(self):
        old_r=int(input("enter old rollno: "))
        new_r=int(input("enter nw rollno: "))
        for i in studentmgmt.students:
            if i.rollno==old_r:
                i.rollno=new_r
                return
            print(f"student with roll no {old_r} not found")
s=studentmgmt('abc',1,24,25)
s.accept()
s.accept()
s.display()
s.search()
s.delete()
s.update()
s.display()
print(studentmgmt.students)
```
