```python
f=open("logfile.txt")
for i in f:
    inhr=int(i.split()[1].split(":")[0][0:2])
    inmin=int(i.split()[1].split(":")[1][0:2])
    outhr=int(i.split()[2].split(":")[0][0:2])
    outmin=int(i.split()[2].split(":")[1][0:2])
    intotal=(inhr*60)+inmin
    outtotal=(outhr*60)+outmin
    final=outtotal-intotal
    if final>=60:
        print(i.split()[0])
f.close()
```

```python
name=input("Enter: ")
f=open("namelist.txt")
l=[i.split() for i in f]
for i in l:
    if len(name)==2: 
        if i[0][0]==name[0] and i[-1][0]==name[1]:
            print(" ".join(i))
    elif len(name)==3:
        if len(i)==3:
            if i[0][0]==name[0] and i[1][0]==name[1] and i[2][0]==name[2]:
                print(" ".join(i))
f.close()
```

```python
f=open("scores.txt")
game=[]
avg=0
for i in f:
    k=i.split()
    game.append([k[1],int(k[2]),k[3],int(k[4])])
for i in range(len(game)):
    avg+=game[i][1]
    avg+=game[i][3]
avg=avg/len(game)
print("1. ")
print(avg)
print()
win=0
loss=0
fav=input("Give fav team name:")
for i in range(len(game)):
    if fav==game[i][0]:
        if game[i][1]>game[i][3]:
            win+=1
        else:
            loss+=1
    elif fav==game[i][2]:
        if game[i][1]<game[i][3]:
            win+=1
        else:
            loss+=1
print("2. ")
print("No. of wins of team ",fav,"= ",win)
print("No. of loss of team ",fav,"= ",loss)
print()
l=[]
for i in range(len(game)):
    if (game[i][1]-game[i][3])>30:
        l.append(game[i][2])
    elif (game[i][3]-game[i][1])>30:
        l.append(game[i][0])
d={}
for i in l:
    d[i]=d.get(i,0)+1
d=sorted(d.items(),key=lambda x:x[-1])
print("3.")
print(d[-1])
print()
d1={}
for i in game:
    d1[i[0]]=d1.get(i[0],[])+[i[1]]
    d1[i[2]]=d1.get(i[2],[])+[i[3]]
print("4. ")
for i,j in d1.items():
    a=sum(j)/len(j)
    if a>=70:
        print(i,a)
print()
f.close()
```

```python
record={}
for team1,score1,team2,score2 in game:
    if team1 not in record:
        record[team1]={'w':0,'l':0,'ms':0,'op':0}
    if team2 not in record:
        record[team2]={'w':0,'l':0,'ms':0,'op':0}
    record[team1]['ms']+=score1
    record[team1]['op']+=score2
    record[team2]['ms']+=score2
    record[team2]['op']+=score1
    if score1>score2:
        record[team1]['w']+=1
        record[team2]['l']+=1
    else:
        record[team2]['w']+=1
        record[team1]['l']+=1
print("5. ")
for i,j in record.items():
    if(j['w']>j['l'])and(j['ms']<j['op']):
        print(i)
```
