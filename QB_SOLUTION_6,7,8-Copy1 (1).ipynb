# 478	Write a python program to read a text file “Story.txt” and print only word starting with ‘I’ in reverse order.

Example: If value in text file is : ‘INDIA IS MY COUNTRY’

Output will be: ‘AIDNI SI MY COUNTRY’


```python
%%writefile Story.txt
INDIA IS MY COUNTRY
is india ?
```

```python
f=open("Story.txt")
l=f.read().split()
for i in l:
    if i[0]=='I' or i[0]=='i':
        print(i[-1::-1],end=" ")
    else:
        print(i,end=" ")
```
